Description:

This bar chart shows the distribution of employees across different teams. The percentage calculation helps understand the workforce balance among teams.

```python
import matplotlib.pyplot as plt

# Count employees in each team
team_distribution = df['Team'].value_counts()

# Calculate percentage
team_percentage = (team_distribution / len(df)) * 100

# Create DataFrame
team_analysis = pd.DataFrame({'Employee Count': team_distribution, 'Percentage (%)': team_percentage})

# Plot the distribution
plt.figure(figsize=(10, 5))
team_distribution.plot(kind='bar', color='skyblue')
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution Across Teams")
plt.xticks(rotation=45)
plt.show()

# Print the result
print(team_analysis)
```

 Description:
This pie chart displays the distribution of employees across different job positions. It visually represents which positions have the highest and lowest number of employees.

```python
position_distribution = df['Position'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(position_distribution, labels=position_distribution.index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen', 'orange'])
plt.title("Employee Distribution by Position")
plt.show()

# Print the result
print(position_distribution)
```

3

 Predominant Age Group Among Employees
 Description:
This bar chart categorizes employees into different age groups and helps identify which group is most common in the company.

```python
bins = [18, 25, 35, 45, 55, 65]
labels = ["18-25", "26-35", "36-45", "46-55", "56-65"]
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_distribution = df['Age Group'].value_counts().sort_index()

# Plot a bar chart
plt.figure(figsize=(8, 5))
age_group_distribution.plot(kind='bar', color='purple')
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.title("Predominant Age Group")
plt.show()

# Print the result
print(age_group_distribution)
```

4️⃣ Team and Position with the Highest Salary Expenditure
 Description:
This analysis helps determine which team and position have the highest salary expenditure, providing insights into where most of the company's salary budget is allocated.

```python
# Total salary by team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False)

# Total salary by position
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False)

# Plot salary expenditure for teams
plt.figure(figsize=(10, 5))
team_salary.plot(kind='bar', color='teal')
plt.xlabel("Team")
plt.ylabel("Total Salary")
plt.title("Salary Expenditure by Team")
plt.xticks(rotation=45)
plt.show()

# Plot salary expenditure for positions
plt.figure(figsize=(8, 5))
position_salary.plot(kind='bar', color='darkred')
plt.xlabel("Position")
plt.ylabel("Total Salary")
plt.title("Salary Expenditure by Position")
plt.xticks(rotation=45)
plt.show()

# Print results
print("Top Teams by Salary Expenditure:\n", team_salary.head())
print("\nTop Positions by Salary Expenditure:\n", position_salary.head())
```

5️⃣ Correlation Between Age and Salary
 Description:
This scatter plot examines the relationship between employee age and salary. A positive correlation would suggest that salary increases with age, while a negative correlation would indicate that younger employees earn more.

```python
import seaborn as sns

# Scatter plot of Age vs. Salary
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Age'], y=df['Salary'], color='blue', alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation Between Age and Salary")
plt.show()

# Compute correlation coefficient
correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary:", correlation)

```
